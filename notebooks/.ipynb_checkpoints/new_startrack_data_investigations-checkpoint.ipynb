```python
# There is a slight difference between the two luminosities calculated by greg in his 2017 and 2019 paper, with the latter ommiting the
# different calculation of Lx for mt < 1. However it is likely that for our bright ULXs all our sources will be mdot > 1 and so the effect
# of this does not matter. 
# HOWEVER if we are seeking to show the entire XLF in our paper, then the luminosities at low mdot should be properly calculated.

plt.xlabel(r'$\dot{m}$')
plt.ylabel(r'$L_{x}$')

L_edd = 1
mdot = np.linspace(0,5, 1000)
Lx = np.where(mdot > 1, L_edd * (1 + np.log(mdot)), L_edd * mdot)
Lx1 = L_edd * (1 + np.log(mdot))
plt.plot(mdot, Lx, label=r'$L_{x}$', c='black')
plt.plot(mdot, Lx1, label=r'$L_{x} = L_{Edd} (1+ln(\dot{m}_{0}))$', c='black', linestyle='--')
plt.axhline(L_edd, label=r'$L_{Edd}$', c='r')

plt.legend()
```

```python
def plot_P_orb_hist(df, cut=10):
    plt.xlabel('Orbital period (days)')
    plt.ylabel('N')
    (df[df['P_orb_days']<cut])['P_orb_days'].hist(bins=100)
```

```python
def plot_system_luminosity_evolution(df_system):
    plt.figure()
    plt.scatter(df_system['t'], np.log10(df_system['Lx1']))
    plt.xlabel('Time')
    plt.ylabel('log (Lx)')
```
