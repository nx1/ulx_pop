```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def load_systems_dataframe(ulx_only=False, beamed=False, half_opening_l_45=False):
    systems_df_path = Path('../data/processed/all_systems_df.csv')
    df = pd.read_csv(systems_df_path)
    if ulx_only:
        df = df[df['Lx'] > 1E39]
    if beamed:
        df = df[df['b'] < 1]
    if half_opening_l_45:
        df = df[df['theta_half_deg'] < 45]
        
    df = df.drop(['Unnamed: 0'], axis=1)
    return df
```

```python
df_all = load_systems_dataframe(ulx_only=False, beamed=False, half_opening_l_45=False)
df_ulx = load_systems_dataframe(ulx_only=True, beamed=False, half_opening_l_45=False)
df_beamed = load_systems_dataframe(ulx_only=True, beamed=True, half_opening_l_45=True)
```

```python
fig, ax = plt.subplots()
plt.title('All Systems')
plt.xlabel('Mass')
plt.ylabel('Frequency')
df_all['m'].hist(ax=ax, bins=100)
ax.set_yscale('log')

fig, ax = plt.subplots()
plt.title('ULX Systems')
plt.xlabel('Mass')
plt.ylabel('Frequency')
df_ulx['m'].hist(ax=ax, bins=100)

fig, ax = plt.subplots()
plt.title('beamed ULX Systems')
plt.xlabel('Mass')
plt.ylabel('Frequency')
df_beamed['m'].hist(ax=ax, bins=100)
```

```python
fig, ax = plt.subplots()
plt.title('All Systems')
plt.xlabel(r'log $\dot{m}_{0}$')
plt.ylabel('Frequency')
df_all = df_all[df_all['mdot_ratio'] < 30000]
np.log10(df_all['mdot_ratio']).hist(ax=ax, bins=100)
ax.set_yscale('log')

fig, ax = plt.subplots()
plt.title('ULX Systems')
plt.xlabel(r'log $\dot{m}_{0}$')
plt.ylabel('Frequency')
df_ulx = df_ulx[df_ulx['mdot_ratio'] < 30000]
np.log10(df_ulx['mdot_ratio']).hist(ax=ax, bins=100)
ax.set_yscale('log')

fig, ax = plt.subplots()
plt.title('beamed ULX Systems')
plt.xlabel(r'log $\dot{m}_{0}$')
plt.ylabel('Frequency')
df_beamed = df_beamed[df_beamed['mdot_ratio'] < 30000]
np.log10(df_beamed['mdot_ratio']).hist(ax=ax, bins=100)
ax.set_yscale('log')
```
