````python
def dict_from_path(path):
    def to_float(val):
        try:
            val = float(val)
        except:
            pass
        return val
    
    p = path
    p = os.path.basename(p)
    p = os.path.splitext(p)[0]
    sp = p.split(',')
    d1 = dict([s.split('-') for s in sp])

    d2 = {k: to_float(v) for k, v in d1.items()}
    return d2
````

````python
hists_cat = np.load('../data/XLF_CAT/XLF_EARNSHAW,N_mc-10000,N_sys-500.npy')
h_cat_mean = np.mean(hists_cat, axis=0)
h_cat_std = np.std(hists_cat, axis=0)
````

````python
csv_paths = glob.glob('../data/XLF/*.npy')


l = []
for csv in csv_paths:
    d = dict_from_path(csv) # Create parameter dict from filename
    h = np.load(csv)   # Read csv file
    d['h'] = h
    l.append(d)
    
df = pd.DataFrame(l)
````

````python
df = df.set_index(['N_mc', 'N_sys', 'Z', 'bh_ratio', 'dincl_max', 'duty_cycle', 'pop_subset', 'xlf_subset', 'L'])
df
````

````python
grid_xlf = {'N_mc'      : [10000],
            'N_sys'     : [500],
            'Z'         : ['all', 0.02, 0.002, 0.0002],
            'bh_ratio'  : [0.0, 0.25, 0.5, 0.75, 1.00],
            'dincl_max' : [46, 21],
            'duty_cycle': [0.2],
            'pop_subset': ['all']} 

# subset of sampled population to create xlfs for
xlf_subsets = ['all', 'ns', 'bh', 'lmxrb', 'alive', 'trans']

# Luminosities to create histograms for
#         Isotropic      Beamed    Beamed (b*d)    precession      precession (d)
grid_L = ['log_Lx_iso', 'log_Lx1', 'log_Lx1_vis', 'log_Lx1_prec', 'log_Lx1_prec_vis']

# For latex plotting
L_dict = {'log_Lx_iso':r'$L_{\mathrm{iso}}$',
          'log_Lx1':r'$L_{\mathrm{x}}$',
          'log_Lx1_vis':r'$L_{\mathrm{x,vis}}$',
          'log_Lx1_prec':r'$L_{\mathrm{x,prec}}$',
          'log_Lx1_prec_vis':r'$L_{\mathrm{x,prec,vis}}$'}
````

````python
# XLF Settings
bin_min = 38
bin_max = 43
bin_width = 0.25
bins = np.arange(bin_min, bin_max, bin_width)
nbins = len(bins)
bin_centers = 0.5 * (bins[:-1] + bins[1:])
````

````python
colors = ['red', 'blue', 'green', 'cyan', 'orange', 'purple']

plt.figure()
fig,ax = plt.subplots(1,len(grid_L), figsize=(25,5), sharey=True)
plt.subplots_adjust(wspace=0)
ax[0].set_ylabel('N(>L)')
ax[0].set_ylim(1,500)
for i, L in enumerate(grid_L):
    c_it = iter(colors)
    for xlf_s in xlf_subsets:
        c = next(c_it)
        hists = df.loc[10000,500,'all',1.0,46,0.2,'all',xlf_s,L]['h']
        h0 = hists[0]
        h_mean = np.mean(hists, axis=0)
        h_std  = np.std(hists, axis=0)
        ax[i].step(bin_centers, h0, color=c)
        ax[i].errorbar(bins[:-1], h_mean, yerr=h_std, capsize=1.0, linewidth=1.0, label=xlf_s, ls='none', color=c)
        ax[i].set_xlabel(L_dict[L])
    ax[i].legend()
    ax[i].set_yscale('log')
    
#plt.savefig('fig.png')
````

This is because the method by Clauset et al 2009 requires unbinned data in order to carry out the maximum likelihood estimate.
------------------

Sampling with replacement from the \textbf{full} binary population
we draw $N_{sys} = 500$ systems realisations with weights proportional to
the sum of their lifetimes during active mass transfer.
We additionally repeat the analysis with the ULX only population
with weights proportional to their time in the ULX phase.

For each of the 500 sampled system, we observe the following luminosities:
$L_{iso}, L_{x}, L_{x,b}, L_{x,b*d}, L_{x,prec}, L_{prec, vis}$
while also seperating out the following populations:
all, NS, BH, LMXB, ALIVE, TRANSIENT.

L_{x,b} corresponds to a random roll against the beaming factor of the system,
assesing whether or not the source was observed down the wind-cone (isotropic emission).
L_{x,b*d} includes the additional roll for LMXB systems to assess whether or not
they were observed during a LMXB outburst.

        L_{x,b} = L_{x} if r < b
                =     0 if r >= b

        L_{x,b*d} = L_{x} if r < b*d
        
                  =     0 if r >= b*d

L_{x,prec} corresponds to the adjusted luminosity obtained from randomly sampling the lightcurve
created via ULXLC. while L_{x,prec,vis} corresponds to the addition of the LMXB duty cycle.

For one sample we thus have 500 systems, with 6 different scenarios that could describe
the luminosity obseved by the source. 

As an aditional form of comparison, we use the catalogue provided in Earnshaw et al, first removing all sources
with luminosity errors EP_8_LUMINOSITY_ERR / EP_8_LUMINOSITY >= 0.8, we than sample the luminosity for 500 sources
assuming assuming guassian errors for the luminosity $ \mu =$ EP_8_LUMINOSITY and $ \sigma =$ EP_8_LUMINOSITY_ERR.
As an aditional form of comparison, we use the catalogue provided in Earnshaw et al, first removing all sources
with luminosity errors EP_8_LUMINOSITY_ERR / EP_8_LUMINOSITY >= 1.0, we than sample the luminosity for 500 sources
assuming assuming guassian errors for the luminosity $ \mu =$ EP_8_LUMINOSITY and $ \sigma =$ EP_8_LUMINOSITY_ERR.

We next perform several fits to both the differential and cumulative luminosity functions.
We make comparisons utilising the catalouge from Earnshaw et al, as well as comparisons
to fits made in the existing literature from a variety of different scenarios.

The cumulative luminosity functions were fit using a  power-law of the form :

$P(L) = CL^{-\alpha}$

re-arranged to the form 
$log(P(L)) = -\alpha log(L) + log(C)$



and the fit was conducted on the tail end of the XLF $(L>10^{39}$ obtained from precession
with and without the inclusion of the LMXB duty cycle, as well as to the XLF obtained from
sampling the ULX catalogue by Earnshaw et al.

The fit was carried out by method of least squares, and the following best fit parameters were obtained:
```

dincl_max = 46 %_BH=0.0  Fit results: L_min39 a = 0.77 +- 0.02 log(c) = 31.68 +- 0.98 (vis)

```



As an aditional form of comparison, we use the catalogue provided in Earnshaw et al, first removing all sources
with luminosity errors EP_8_LUMINOSITY_ERR / EP_8_LUMINOSITY >= 1.0, we than sample the luminosity for 500 sources
assuming assuming guassian errors for the luminosity $ \mu =$ EP_8_LUMINOSITY and $ \sigma =$ EP_8_LUMINOSITY_ERR.

we perform fits using different $L_{\mathrm{min}}$starting at each order of magnitude between
L=10^36 - 10^40. We fit the data using the method of maximum likelihood described in Clauset et al (2007)
and obtain 1 sigma errors on the the slope $\alpha$ via 1000 Monte-Carlo iterations.


```
QUOTE EARNSHAW 2018:
The error on the luminosity was  derived  from  the  error  on  the  flux,  as  while  we  expect  that the error
will be dominated by the uncertainty in the distance measurement, this is likely dominated by unknown systematics
and so is not well quantified. We defined ULXs as those sources with LXâ‰¥1039 erg sâˆ’1 or luminosity within 1Ïƒ of this
value, however, we also retain a much larger number of lower luminosity sources in our  sample  as  an  extension 
of  the  catalogue.  We  note  that  these are all apparent luminosities as we do not make any corrections for absorption.
```

````python
def hists_from_Ls(Ls, bins, cumulative=False):
    nbins = len(bins)
    hs = np.ndarray((len(Ls), nbins-1)) # array for storing histograms
    for i, L in enumerate(Ls):    
        h = np.histogram(np.log10(L), bins=bins)[0]
        if cumulative:
            h_cum = np.cumsum(h)
            h_surv = sum(h) - h_cum
            hs[i] = h_surv
            #assert(h.max() == N_sys) # Force full sample in hist    
        else:
            hs[i] = h
    return hs


# XLF Settings
bin_min = 33
bin_max = 43
bin_width = 0.2
bins = np.arange(bin_min, bin_max, bin_width)
nbins = len(bins)
bin_centers = 0.5 * (bins[:-1] + bins[1:])

Ls = df_L.loc[1000,500,'all',0.5,46,0.2,'all','ns','Lx1_prec']['Ls']
hs = hists_from_Ls(Ls, bins=bins, cumulative=False)

mu = np.mean(hs, axis=0)
std = np.std(hs, axis=0)

plt.figure(figsize=(6,6))
plt.ylabel('dN/dL')

plt.errorbar(bin_centers, mu, yerr=std, linestyle='None')
plt.step(bin_centers, mu, where='mid')
plt.yscale('log')
plt.ylim(0,600)
````

````python
def log_line(logx, a, logc):
    return -a*logx+logc

def fit_loglog(log_x_min, logx, y, ystd):
    df = pd.DataFrame()
    df['y'] = y
    df['ystd'] = ystd
    df['logx'] = logx
    df['logy'] = np.log10(y)
    df['logyerr'] = 0.43429 * ystd / y

    df = df[df['logy'] > 0]
    df_fit = df[df['logx'] > log_x_min]
    
    p_opt, p_cov = curve_fit(log_line, xdata=df_fit['logx'], ydata=df_fit['logy'], sigma=df_fit['logyerr'], maxfev=10000)
    perr = np.sqrt(np.diag(p_cov))
    print(f'Fit results: L_min={log_x_min} a = {p_opt[0]:.2f} +- {perr[0]:.2f} log(c) = {p_opt[1]:.2f} +- {perr[1]:.2f}')
    return df_fit, p_opt, p_cov
````

A cutoff powerlaw of the form

$ N(>L) = A (L^{-\alpha} - L_{max}^{-\alpha}) $

is given in Grimm (2003) thesis

````python
def cutoff_pl(L, L_max, alpha, A):
    return A * (L**-alpha - L_max**-alpha)
````
