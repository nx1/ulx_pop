```python
def dict_from_path(path):
    def to_float(val):
        try:
            val = float(val)
        except:
            pass
        return val
    
    p = path
    p = os.path.basename(p)
    p = os.path.splitext(p)[0]
    sp = p.split(',')
    d1 = dict([s.split('-') for s in sp])

    d2 = {k: to_float(v) for k, v in d1.items()}
    return d2
```

```python
csv_paths = glob.glob('../data/XLF/*.npy')


l = []
for csv in csv_paths:
    d = dict_from_path(csv) # Create parameter dict from filename
    h = np.load(csv)   # Read csv file
    d['h'] = h
    l.append(d)
    
df = pd.DataFrame(l)
```
